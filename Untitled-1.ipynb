 下载次数和发表人关系

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import expit  # 计算Logistic函数
import datetime

r_f=pd.read_csv(r"C:\Users\宋志坤\Desktop\新建文件夹\学习\新建文件夹\qwq.csv")

holder_loadnum=r_f.groupby("Holder")["下载次数"].sum()
print(holder_loadnum)
sns.boxplot(x='Holder', y='下载次数', data=r_f)
plt.xticks(rotation=45)
plt.title('附件下载次数与通知人关系')
plt.show()
```

 日期复杂度

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import expit  # 计算Logistic函数
import datetime

r_f=pd.read_csv(r"C:\Users\宋志坤\Desktop\新建文件夹\学习\新建文件夹\qwq.csv")
r_f['Date']=pd.to_datetime(r_f['Date'])
r_f["Date"]=r_f["Date"].dt.date
time_series=r_f.groupby("Date").size()
plt.figure(figsize=(12, 6))
time_series.plot(kind='line', color='b')
plt.title('每日通知数')
plt.xlabel('日期')
plt.ylabel('通知数')
plt.show()
```

 年份

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import expit  # 计算Logistic函数
import datetime

r_f=pd.read_csv(r"C:\Users\宋志坤\Desktop\新建文件夹\学习\新建文件夹\qwq.csv")
r_f['Date']=pd.to_datetime(r_f['Date'])
r_f["Date"]=r_f["Date"].dt.year
year_series=r_f.groupby("Date").size()
print(year_series)
```

 logistic函数

```python
x=np.linspace(-10,10,400)
y=1/(1+expit(-1*x))
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
